```
from fastbook import *
from fastai.vision.widgets import *
import pandas as pd
import numpy as np
import joblib
import sklearn

# !pip install voila
# !jupyter serverextension enable --sys-prefix voila
```

```
rank1_upload = widgets.IntText(
    value = '10',
    description='Rank1:',
    disabled=False
)
rank1_upload

rank2_upload = widgets.IntText(
    value = '10',
    description='Rank2:',
    disabled=False
)
rank2_upload

elo1_upload = widgets.IntText(
    value = '10',
    description='Elo1:',
    disabled=False
)


elo2_upload = widgets.IntText(
    value = '10',
    description='Elo2:',
    disabled=False
)


C_upload = widgets.BoundedIntText(
    value=0,
    min=0,
    max=1,
    step=1,
    description='C 1 or 0',
    disabled=False
)

G_upload = widgets.BoundedIntText(
    value=0,
    min=0,
    max=1,
    step=1,
    description='H 1 or 0',
    disabled=False
)

In_upload = widgets.BoundedIntText(
    value=0,
    min=0,
    max=1,
    step=1,
    description='Out 1 or 0',
    disabled=False
)
```

```
# load model
filename = '/content/money_machine_V_01.sav'
loaded_model = joblib.load(filename)
```

```
# input_array = [[rank1_upload.value, rank2_upload.value,elo1_upload.value,elo2_upload.value, C_upload.value, G_upload.value, In_upload.value]]
```

```
btn_run = widgets.Button(description='Predict')
def on_click_classify(change):
    input_array = [[rank1_upload.value, rank2_upload.value,elo1_upload.value,elo2_upload.value, C_upload.value, G_upload.value, In_upload.value]]
    result = loaded_model.predict(input_array)
    print(result)
    return result

btn_run.on_click(on_click_classify)
VBox([widgets.Label('Select your ods!'), 
      rank1_upload, rank2_upload, elo1_upload, elo2_upload, C_upload,G_upload, In_upload,  btn_run])
```
